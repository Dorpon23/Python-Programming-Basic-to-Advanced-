```python
import folium
map = folium.Map(location=[59.349849, 18.070257],tiles="Stamen Terrain")
fg = folium.FeatureGroup(name="My Map")
for coordinates in [[59.349849, 18.070257],[53.2, 15.3]]:
    fg.add_child(folium.Marker(location=coordinates,popup="Your Position",icon=folium.Icon(color='green')))
map.add_child(fg)
```

```python
import folium
map = folium.Map(location=[38.58, -99.09],zoom_start=15,tiles="Stamen Terrain")
fg = folium.FeatureGroup(name="My Map")
for coordinates in [[38.2, -99.2],[39.2, -97.1]]:
    fg.add_child(folium.Marker(location=coordinates,popup="Your Position",icon=folium.Icon(color="green")))
map.add_child(fg)
```

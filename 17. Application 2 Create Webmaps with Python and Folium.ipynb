```python
import folium
import pandas
data = pandas.read_csv("C:\\Users\\Dorpon\\Python\\1.1 app2-web-map\\volcanoes.txt")
lat = list(data["LAT"])
lon = list(data["LON"])
nam = list(data["NAME"])
elv = list(data["ELEV"])

def color_proceducer(elevation):
    if elevation < 1800:
        return "green"
    elif 1800<=elevation <3000:
        return "blue"
    else:
        return "red"
    
map = folium.Map(location=[39.37568019221214, -101.40882019892456])

fg = folium.FeatureGroup(name="My Map")
for co,co1,na,el in zip(lat,lon,nam,elv):
    fg.add_child(folium.Marker(location=[co,co1],popup= na,icon=folium.Icon(color=color_proceducer(el))))
map.add_child(fg)
map.save("volcano.html")
#zip function documentation: https://www.w3schools.com/python/ref_func_zip.asp
```

```python
import folium
import pandas
data = pandas.read_csv("C:\\Users\\Dorpon\\Python\\1.1 app2-web-map\\volcanoes.txt")
lat = list(data["LAT"])
lon = list(data["LON"])
nam = list(data["NAME"])
elv = list(data["ELEV"])

def color_proceducer(elevation):
    if elevation < 1800:
        return "green"
    elif 1800<=elevation <3000:
        return "blue"
    else:
        return "red"
    
map = folium.Map(location=[39.37568019221214, -101.40882019892456])

fg = folium.FeatureGroup(name="My Map")
for co,co1,na,el in zip(lat,lon,nam,elv):
    fg.add_child(folium.CircleMarker(location=[co,co1],popup= na,radius=6,fill_collor=color_proceducer(el),
                                    fill_opacity=0.7))
map.add_child(fg)
map.save("volcanocircale.html")
#zip function documentation: https://www.w3schools.com/python/ref_func_zip.asp
```

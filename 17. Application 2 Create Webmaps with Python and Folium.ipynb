```python
import folium
map = folium.Map(location=[59.349849, 18.070257],tiles="Stamen Terrain")
fg = folium.FeatureGroup(name="My Map")
for coordinates in [[59.349849, 18.070257],[53.2, 15.3]]:
    fg.add_child(folium.Marker(location=coordinates,popup="Your Position",icon=folium.Icon(color='green')))
map.add_child(fg)
```

```python
import folium
import pandas
data = pandas.read_csv("C:\\Users\\Dorpon\\Python\\1.1 app2-web-map\\volcanoes.txt")
lat = list(data["LAT"])
lon = list(data["LON"])
#la = len(lat)
#lo =len(lon)
fg = folium.FeatureGroup(name="My Map")
for coordinates in zip(lat, lon):
    fg.add_child(folium.Marker(location=coordinates,popup="Your Position",icon=folium.Icon(color='green')))
map.add_child(fg)
map.save("")
```
